```python
import mysql.connector
import pandas as pd
from IPython.display import display,HTML

#reemplazar 'root' por el password del usuario administrador de MySQL
db_connection = mysql.connector.connect(user="root",host="localhost",password="racing11")
cursor = db_connection.cursor()
cursor.execute("USE sakila")
```

1) A partir de la tabla "film", mostrar solamente cuáles son los títulos de pelicula disponibles.

```python
df = pd.read_sql('SELECT title FROM film limit 10;', con=db_connection)
display(df)
```

2) A partir de la tabla "film", mostrar de qué años son las peliculas disponibles. NB: utilizar la instrucción DISTINCT

```python
df = pd.read_sql('SELECT DISTINCT release_year FROM film;', con=db_connection)
display(df)
```

3) Mostrar los títulos de pelicula en la tabla "film" que se puede arrendar por más de 4 días (rental_duration)

```python
df = pd.read_sql(' SELECT title FROM film WHERE rental_duration>3;', con=db_connection)
display(df)
```

4) En la tabla "payment", mostrar el customer_id, rental_id, amount y payment_date de todos los arriendos realizados entre las fechas 2005-07-01 y 2005-08-01

1) Contar el número de clientes en la tabla "CUSTOMER"

```python
df = pd.read_sql('SELECT COUNT(*) FROM customer;', con=db_connection)
display(df)
```

2) En la tabla "RENTAL", contar cuánto arriendos se realizó por cada cliente (NB: utilizar la instucción GROUP BY y COUNT)

```python
df = pd.read_sql('SELECT customer_id, COUNT(*) FROM  rental GROUP BY customer_id;', con=db_connection)
display(df)
```

3) ¿Qué devuelve la consulta SQL siguiente?

```python
df = pd.read_sql('SELECT MAX(total) FROM (SELECT customer_id,count(*) as total FROM rental GROUP BY customer_id) as t1;', con=db_connection)
display(df)
```

4) ¿Qué es un sub-consulta?

Una subconsulta es una instrucción SELECT anidada dentro de una instrucción SELECT, SELECT...INTO, INSERT...INTO, DELETE, o UPDATE o dentro de otra subconsulta.

Puede utilizar tres formas de sintaxis para crear una subconsulta:

   <code> comparación [ANY | ALL | SOME] (instrucción sql)
    expresión [NOT] IN (instrucción sql)
    [NOT] EXISTS (instrucción sql) <code>
    
 comparación

Es una expresión y un operador de comparación que compara la expresión con el resultado de la subconsulta.

expresión

Es una expresión por la que se busca el conjunto resultante de la subconsulta. 

5) En la tabla "payment", mostrar el volumen de negocio realizado por cada miembro del staff (staff_id) en total.

```python
df = pd.read_sql('SELECT staff_id, sum(amount) from payment group by staff_id;', con=db_connection)
display(df)
```

6) En una consulta, mostrar el staff_id que ha realizado más volumen de negocio el dia 2005-08-23.

```python
df = pd.read_sql('SELECT staff_id from payment where payment_date=\'2005-08-23\' group by staff_id order by sum(amount) desc limit 1;', con=db_connection)
display(df)
```
